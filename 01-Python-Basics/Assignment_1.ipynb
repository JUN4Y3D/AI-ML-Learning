```python

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2 if num2 != 0 else "undefined (cannot divide by zero)"

print(f"Sum: {sum}")
print(f"Difference: {difference}")
print(f"Product: {product}")
print(f"Quotient: {quotient}")

print(type(sum))
print(type(difference))
print(type(product))
print(type(quotient))
```

```python
num1 = 10
num2 = 5

tmp = num1
num1 = num2
num2 = tmp
print(f"After swapping: num1 = {num1}, num2 = {num2}")

# Alternatively, using tuple unpacking
num1, num2 = num2, num1
print(f"After swapping again using tuple unpacking: num1 = {num1}, num2 = {num2}")

# Swapping without a temporary variable
num1 = num1 + num2
num2 = num1 - num2
num1 = num1 - num2
print(f"After swapping again without temporary variable: num1 = {num1}, num2 = {num2}")

# Using XOR bitwise operation to swap
num1 = num1 ^ num2
num2 = num1 ^ num2
num1 = num1 ^ num2
print(f"After swapping again using XOR: num1 = {num1}, num2 = {num2}")
```

```python
num = float(input("Enter a number: "))

integer_part = int(num)
fractional_part = num - integer_part
print(f"Integer part: {integer_part}")
print(f"Fractional part: {fractional_part}")

#print only fractional part, first two digits after decimal point
fractional_part_two_digits = round(fractional_part, 2) # rounding to 2 decimal places
print(f"Fractional part (two digits): {fractional_part_two_digits}")
```
