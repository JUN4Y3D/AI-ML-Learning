```python
salary = int(input("Enter your salary: "))

if(salary<30000):
    print("Your tax is 5%")
elif(salary>=30000 and salary<=70000):
    print("Your tax is 15%")
elif(salary>70000):
    print("Your tax is 25%")
```

```python
def pnt(a,b):
    for i in range(a,b+1):
        if(not(i&1)):
            print(i, end=' ')

pnt(1,10)
```

```python
n=int(input("Enter a num: "))
rev=0
while(n!=0):
    digit=n%10
    rev=rev*10+digit
    n=n//10

while(rev!=0):
    digit=rev%10
    print(digit, end=' ')
    rev=rev//10
```

```python
#total num of digit
def count(n):
	cnt = 0
	while(n!=0):
		cnt+=1
		n//=10
	return cnt
print(count(1200030))
```

```python
while True:
    value = input("Enter a number (or type Quit to stop): ")

    if value.lower() == "quit":      # stop condition
        break

    num = float(value)       # convert input to number

    if num > 0:
        print("Positive")
    elif num < 0:
        print("Negative")
    else:
        print("Zero")
```

```python
def calculator(a,b,s):
    if s == '+':
        print(a+b)
    elif s == '-':
        print(a-b)
    elif s == '*':
        print(a*b)
    elif s == '/':
        if b == 0:
            print("Division by zero is not allowed")
        else:
            print(a / b)

calculator(10,2,'/')
```
