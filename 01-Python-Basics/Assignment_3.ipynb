```python
s = "abcd"
n = len(s)
fg = True

for i in range(n//2):
	if s[i]!=s[n-i-1]:
		fg = False
		break

if(fg):
	print("Palindrome")
else:
	print("Not a Palindrome")

#different method

s1 = "acca"
x = s1[::-1] # s[start : end : step] , Start from the end (because step = -1) , Move backwards by 1 , Go until the beginning

if(s1==x):
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
info = {}   # Dictionary to store student name → marks

while True:
    print("\n--- Menu ---")
    print("A - Add a student")
    print("B - Update marks")
    print("C - Search for a student")
    print("D - Display all students")
    print("E - Exit")

    choice = input("Enter your choice: ").upper()

    if choice == "A":
        name = input("Enter student name: ")
        marks = int(input("Enter marks: "))
        info[name] = marks
        print("Student added successfully!")

    elif choice == "B":
        name = input("Enter student name to update: ")
        if name in info:
            marks = int(input("Enter new marks: "))
            info[name] = marks
            print("Marks updated!")
        else:
            print("Student not found!")

    elif choice == "C":
        name = input("Enter student name to search: ")
        if name in info:
            print(f"{name} → {info[name]} marks")
        else:
            print("Student not found!")

    elif choice == "D":
        if len(info) == 0:
            print("No students available.")
        else:
            print("\n--- Student List ---")
            for name, marks in info.items():
                print(f"{name} : {marks}")

    elif choice == "E":
        print("Exiting program...")
        break

    else:
        print("Invalid choice! Please enter A, B, C, D, or E.")
```

```python
lst = [1, 2, 3, 2, 4, 5, 1, 6, 3]

seen = set()        # to store elements we have seen
duplicates = set()  # to store elements that appear more than once

for x in lst:
    if x in seen:
        duplicates.add(x)  # already seen → it's a duplicate
    else:
        seen.add(x)        # first time → add to seen

print("Elements that appear more than once:", duplicates)
```
